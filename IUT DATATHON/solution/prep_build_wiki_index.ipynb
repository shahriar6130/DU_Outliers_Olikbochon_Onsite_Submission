Source: `wikimedia/wikipedia`, config `20231101.bn` (CC BY-SA). Cite Bengali Wikipedia in the paper.


```python
%pip install -q datasets
from datasets import load_dataset
import pandas as pd, re

ds = load_dataset("wikimedia/wikipedia", "20231101.bn", split="train")
print(len(ds), "articles")
```

```python
rows = []
for a in ds:
    title = a["title"]
    paras = [p.strip() for p in a["text"].split("\n") if len(p.strip()) >= 120]
    for p in paras[:3]:  # lead paragraphs carry the factoid density
        rows.append({"title": title, "text": f"{title}: {p[:900]}"})

df = pd.DataFrame(rows)
print(len(df), "passages")
df.to_parquet("bn_wiki_passages.parquet", index=False)
print("saved bn_wiki_passages.parquet — upload as a PUBLIC Kaggle dataset and attach to the inference notebook")
```
